```python
def education_to_numeric(x):  # education will also need to be numeric.
    if x == "Bachelor's":
        return 7
    if x == "Master's":
        return 8
    if x == "High School":
        return 9
    if x == "Doctorate":
        return 10


data["educationNum"] = data["education_of_employee"].apply(education_to_numeric)
print(data)
```
